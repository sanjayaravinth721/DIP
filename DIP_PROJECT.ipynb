```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img = Image.open("/content/pexels-helena-lopes-2253275.jpg")
```

```python
img
```

```python
img.show()
```

```python
plt.imshow(img)
```

```python
print(img.size)
print(img.format)
print(img.mode)
```

```python
img.save("Dog_from_python.jpg")
```

```python
#cropping
left=2000
right=1000
top=1000
bottom=3000

crop_img = img.crop((left,top,right,bottom))
plt.imshow(crop_img)
```

```python
#transpose

transpose_right = img.transpose(Image.FLIP_LEFT_RIGHT)
plt.imshow(transpose_right)
plt.title("From right")
plt.subplot(3,2,2)
```

```python
#rotate

plt.imshow(img.rotate(180))
```

```python
#resized
newsize=(300,300)
fig_size=(10,10)
plt.figure(figsize=fig_size)
r1 = img.resize(newsize,Image.BILINEAR)
r2 = img.resize(newsize,Image.NEAREST)
r3 = img.resize(newsize,Image.BICUBIC)
r4 = img.resize(newsize,Image.HAMMING)
plt.subplot(3,3,1)
plt.title("bilinear")
plt.imshow(r1)
plt.subplot(3,3,2)
plt.title("Nearest")
plt.imshow(r2)
plt.subplot(3,3,3)
plt.title("Bicubic")
plt.imshow(r3)
plt.subplot(3,3,4)
plt.title("Hamming")
plt.imshow(r4)


```

```python
#image watermark

size=(500,500)
crop_image = img.copy()
crop_image.thumbnail(size)

crop_image2 = img.copy()
crop_image2.paste(crop_image,(1000,1000))

plt.imshow(crop_image2)
```

```python
#blackANDwhite

bw_image = img.convert("L")
plt.imshow(bw_image,cmap='gray')
```

```python
#image to numpy array

numpy_array = np.array(img)
print(numpy_array.shape)
```

```python
#numpy to image

num_img = Image.fromarray(numpy_array)
plt.imshow(num_img)
```
